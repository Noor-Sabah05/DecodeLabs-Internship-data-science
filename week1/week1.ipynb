## Task 3: Exploratory Data Analysis (EDA) 


```python
import pandas as pd
df1 = pd.read_excel("../datasets/week1_cleaned.xlsx")
```

```python
df1.columns
```

```python
df1.info()
```

```python
df1.describe()
```

as we can see the missing values are removed. The data columns are right skewed, means the std is less than the mean which means few outliers pull the mean high. We will try to remove outliers from the data and make dist normal. 

```python
num_cols = ["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]
```

```python
df1[num_cols].agg(["mean", "median", "std", "var", "skew", "kurt"])
#> 0  → right skew, < 0  → left skew, ≈ 0  → symmetric
```

we can see that the totalPrice column is highly right skewed, which means there are few outliers that pull the mean high. We can try to remove outliers from the data and make the distribution normal.

```python
# product analysis to understand which products are most popular and generate the most revenue. This can help in inventory management and marketing strategies.
product_quantity= ( df1.groupby("Product")["Quantity"].sum().sort_values(ascending=False))

print(product_quantity)
```

```python
product_revenue = df1.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)

print(product_revenue)
```

```python
customer_orders= ( df1['CustomerID'].value_counts().sort_values(ascending=False))
customer_orders.head(10)
```

```python
customer_orders.describe()
```

```python
corr = df1[num_cols].corr()

print(corr)
```

```python
# outliers finding using IQR method
for col in num_cols:

    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df1[(df1[col] < lower) | (df1[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

```python
outliers["TotalPrice"]
```

```python
df2=df1.drop(outliers.index, inplace=False)
```

```python
df2["TotalPrice"].mean()
```

u can see how much the 8 outliers in total price were effecting the mean , removal of them took mean down from so high to avg value 

```python
df2["TotalPrice"].std()
```

```python
df2.to_excel("../datasets/week1_outliers_removed.xlsx", index=False)
```
